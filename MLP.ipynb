# Multilayer Perceptron

## Pam50 dataset

```python
import pandas as pd

### Read csv file
Data_Pam = pd.read_csv("data/Data_Pam50.csv")
Data_Pam
```

### Splitting file into label and data

```python
label_raw = Data_Pam.iloc[:,1]
data_raw = Data_Pam.iloc[:,2:]
label_raw
```

```python
from sklearn.model_selection import train_test_split
from collections import Counter
X_train, X_test, y_train, y_test = train_test_split(data_raw, label_raw,random_state=1, test_size=0.2, stratify = label_raw)
print("y_train: ", Counter(y_train))
print("y_test: ", Counter(y_test))
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_trainscaled=sc_X.fit_transform(X_train)
X_testscaled=sc_X.transform(X_test)
```

### Model

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(hidden_layer_sizes=(256,128,64,32),activation="relu",random_state=1).fit(X_trainscaled, y_train)
y_pred=clf.predict(X_testscaled)
print(clf.score(X_testscaled, y_test))
y_pred
```

## Multidrug dataset

```python
### Read csv file
Data_multidrug = pd.read_csv("data/multidrug_data.csv", sep='"/\t/"', engine='python')
Data_multidrug
```

```python
Data_multidrug=Data_multidrug.drop(Data_multidrug.columns[0], axis=1)
Data_multidrug
```

```python
Data_multidrug=Data_multidrug.drop(labels=6, axis=0)
Data_multidrug
```

```python
Data_multidrug['ABCG8"/'] = Data_multidrug['ABCG8"/'].str.replace('"/',"")
Data_multidrug
```

```python
label_raw = Data_multidrug.iloc[:,0]
data_raw = Data_multidrug.iloc[:,1:]
label_raw
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_raw, label_raw,random_state=1, test_size=0.3, stratify = label_raw)
print("y_train: ", Counter(y_train))
print("y_test: ", Counter(y_test))
```

```python
sc_X = StandardScaler()
X_trainscaled=sc_X.fit_transform(X_train)
X_testscaled=sc_X.transform(X_test)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(256,128,64,32),activation="relu",random_state=1).fit(X_trainscaled, y_train)
y_pred=clf.predict(X_testscaled)
print(clf.score(X_testscaled, y_test))
y_pred
```
